```python
from scipy import stats
import numpy as np

#One-sample t-test
sample_data = np.array([10, 12, 11, 13, 10, 9, 12, 11, 10, 13])
population_mean = 10.5
t_statistic, p_value = stats.ttest_1samp(sample_data, population_mean)
print(f"one-sample t-test: t-statistic = {t_statistic:.2f}, p-value = {p_value:.3f}")

```

```python
#Two-sample independent t-test
group1 = np.array([25, 28, 27, 26, 29])
group2 = np.array([22, 24, 23, 21, 25])
t_statistic, p_value = stats.ttest_ind(group1, group2)
print(f"two-sample independent t-test: t-statistic = {t_statistic:.2f}, p-value = {p_value:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#Generate example data for two groups
np.random.seed(42)
group_a = np.random.normal(loc=50, scale=10, size=100)
group_b = np.random.normal(loc=55, scale=10, size=100)

#Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(group_a, group_b, equal_var=True)

#create a figure and axes for the plot
plt.figure(figsize=(10, 6))

#plot histograms of the two groups
sns.histplot(group_a, color='skyblue', label='Group A', kde=True, stat='density',alpha=0.6)
sns.histplot(group_b, color='salmon', label='Group B', kde=True, stat='density',alpha=0.6)

#Add vertical line for the mean of each group
plt.axvline(np.mean(group_a), color='blue', linestyle='--',linewidth=2, label=' Mean Group A')
plt.axvline(np.mean(group_b), color='red', linestyle='--', linewidth=2,  label='Mean Group B ')

#
```
